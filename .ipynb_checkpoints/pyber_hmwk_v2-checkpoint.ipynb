## Pyber Ride Sharing Observations
- The number of rides taken in urban areas is significantly higher than the number of rides taken in the rural     areas.
- The average fare in the rural areas tends to be greater than the average fare in the urban areas. One might     assume that the distances between origination and destination might be larger in the rural areas, resulting       in the higher average fare.
- The data indicates that urban areas account for over 60% of total fares(62.7%) and total rides(68.4%) for       2018.
- The urban areas have more than four times the number of drivers than the rural and suburban areas combined.     Driver competition might account for the lower average fares in the urban areas as a result. 

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)

ride_data_df = pd.read_csv(ride_data_to_load)


city_data_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
rsd_df = city_group['fare'].mean()
rsd_df = pd.DataFrame(rsd_df)
rsd_df['ride_count'] = pd.DataFrame(city_group['ride_id'].count())
rsd_df['driver_count'] = city_group["driver_count"].max()
rsd_df = rsd_df.rename(columns={'fare':'Avg Fare'})
rsd_df.head()
# Obtain the x and y coordinates for each of the three city types
rural_x = rsd_df.loc["Rural", "ride_count"].values
rural_y = rsd_df.loc["Rural", "Avg Fare"].values
rural_z = rsd_df.loc["Rural", "driver_count"].values

suburban_x = rsd_df.loc["Suburban", "ride_count"].values
suburban_y = rsd_df.loc["Suburban", "Avg Fare"].values
suburban_z = rsd_df.loc["Suburban", "driver_count"].values

urban_x = rsd_df.loc["Urban", "ride_count"].values
urban_y = rsd_df.loc["Urban", "Avg Fare"].values
urban_z = rsd_df.loc["Urban", "driver_count"].values

# Build the scatter plots for each city types
ruralplt = plt.scatter(rural_x, rural_y, s=rural_z*10, c="gold", marker="o", edgecolors="black", alpha=".75", label="Rural")
suburbanplt = plt.scatter(suburban_x, suburban_y, s=suburban_z*10, c="lightskyblue", marker="o", edgecolors="black", alpha=".75", label="Suburban")
urbanplt = plt.scatter(urban_x, urban_y, s=urban_z*10, c="lightcoral", marker="o", edgecolors="black", alpha=".75", label="Urban")


# Incorporate the other graph properties
plt.grid(True)
plt.title("Pyber Ride Sharing Data (2018)", fontsize=15)
plt.xlabel("Total Number of Rides (Per City)", fontsize=12)
plt.ylabel("Average Fare ($)", fontsize=12)

# Create a legend
lgnd = plt.legend(loc="best", title="City Type", title_fontsize=12, fontsize=10, markerscale=.5, facecolor="white")

# Incorporate a text label regarding circle size
plt.text(45, 35, 'Note:\nCircle size correlates with driver count per city', fontsize=12)

# Save Figure
plt.savefig("PyberRideSharingData2018.png")
plt.show()
```

```python
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0, 0 , .1)
```

## Total Fares by City Type

```python
# Calculate Type Percents
ride_type_df = pd.DataFrame(combined_df)
ride_type_group = ride_type_df.groupby(["type"])
ride_type_breakdown = pd.DataFrame({"Total Count":ride_type_group["type"].count(),"Percentage of Total Rides":ride_type_group["type"].count()/ride_count*100})
# Calculate Ride Percents
ride_fare_df = pd.DataFrame(combined_df)
ride_fare_group = ride_fare_df.groupby(["type"])
ride_fare_breakdown = pd.DataFrame({"Total Spend":ride_fare_group["fare"].sum(),"Percentage of Total Spend":ride_fare_group["fare"].sum()/total_fare*100})

ride_fare_breakdown.head()
```

```python
# Build Pie Chart
plt.pie(ride_fare_breakdown["Percentage of Total Spend"], labels = ride_type_breakdown.index, explode=explode, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
# Create axes which are equal so we have a perfect circle
plt.axis("auto")

plt.title("% of Total Fares by City Type")

plt.show()
# Save Figure
plt.savefig("Pyber_TotFare_CityType.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ride_type_df = pd.DataFrame(combined_df)
ride_type_group = ride_type_df.groupby(["type"])
ride_type_breakdown = pd.DataFrame({"Total Count":ride_type_group["type"].count(),"Percentage of Total Rides":ride_type_group["type"].count()/ride_count*100})

ride_type_breakdown.head()
```

```python
# Build Pie Chart
plt.pie(ride_type_breakdown["Percentage of Total Rides"], labels = ride_type_breakdown.index, explode=explode, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
# Create axes which are equal so we have a perfect circle
plt.axis("auto")

plt.title("% of Total Rides by City Type")

plt.show()
# Save Figure
plt.savefig("Pyber_TotRides_CityType.png")
```
